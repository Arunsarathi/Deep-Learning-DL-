```python
# Convolution Neural Network on MNIST handwritten dataset
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Convolution2D, MaxPooling2D
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('mnist.csv')
```

```python
data.head()
```

```python
df_x = data.iloc[:,1:].values.reshape(len(data),28,28,1)
y = data.iloc[:,0].values
```

```python
# converting label to categorical feature

df_y = keras.utils.to_categorical(y,num_classes=10)
```

```python
df_x = np.array(df_x)
df_y = np.array(df_y)
```

```python
df_y
```

```python
df_x.shape
```

```python
df_y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_x,df_y, test_size=0.2, random_state=4)
```

```python
# CNN model
model =Sequential()
model.add(Convolution2D(32,3,data_format='channels_last',activation='relu',input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(100))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy',
              optimizer='adadelta',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,nb_epoch=3, validation_data=(x_test,y_test))
```

```python
model.evaluate(x_test,y_test)
```
